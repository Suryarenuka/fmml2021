1. How do you think would the number of neighbors effect the Isomap algorithm? What happens when the number of neighbors considered is very large? What happens when it is very low?

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
```

```python
n_points = 1000
X, color = datasets.make_s_curve(n_points, random_state=0)
```

```python
fig = plt.figure(figsize=(15, 12))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.Spectral)
ax.view_init(4, -72)
```

```python
from sklearn.manifold import Isomap
```

```python
Y = Isomap(n_neighbors=4, n_components=2).fit_transform(X)

ax = fig.add_subplot(111)
plt.scatter(Y[:, 0], Y[:, 1], c=color, cmap=plt.cm.Spectral)
plt.title("Isomap on S curve")
plt.axis('off')
plt.show()
```

Answer

1.The manifold embedding result is generally highly dependent on the number of neighbors chosen.

2.In manifold learning, as the number of neighbors k value increases data points may 'short cirucit' the manifold and drastically change the embedding.

3.In manifold learning, if we decrease number of neighbors k value decreases to lowest point the distance between points turn out ot be "infinity" and results into two disjoint manifolds.

4.There should exist a pact for every point and the graph that we inspected from the data points should fully connected. this happens only when k value ideal means not too large or small.
